```python
from sentence_transformers import SentenceTransformer, util

# Load the pre-trained model
sentence_similarity = SentenceTransformer('symanto/sn-xlm-roberta-base-snli-mnli-anli-xnli')
```

```python
from tqdm import tqdm

def encode_question(question):
    # Encode the question using the SentenceTransformer model and return the resulting tensor
    return sentence_similarity.encode(question, convert_to_tensor=True)

# Apply the encoding function to the "question" column and create a new column with the encoded tensors
tqdm.pandas(desc="Encoding questions")
df['question_embeddings'] = df['question'].progress_apply(encode_question)
```

```python
import torch

def convert_to_list(tensor):
  return tensor.tolist()

df['question_embeddings'] = df['question_embeddings'].apply(convert_to_list)
```

```python
import torch
import ast

def findSimilarity(source_sentence):

  target_sentences = df['question'].tolist()
  source_embedding = sentence_similarity.encode(source_sentence, convert_to_tensor=True)

  similarity_scores = util.pytorch_cos_sim(source_embedding, torch.tensor(
        df['question_embeddings'].apply(ast.literal_eval), device='cpu:0'))

  # Sort the target sentences by their similarity score
  sorted_indices = similarity_scores.argsort(descending=True)[0]
  hightest_score = similarity_scores[0][sorted_indices[0]].item()
  sorted_targets = [target_sentences[i] for i in sorted_indices]
  return df[df['question'] == sorted_targets[0]]['answer'].tolist()[0], hightest_score
```

```python
import nltk
from nltk.chat.util import Chat, reflections
from transformers import Conversation, EncoderDecoderModel, AutoTokenizer,pipeline

# Load the tokenizer and model
#https://github.com/aub-mind/Arabic-Empathetic-Chatbot
#https://huggingface.co/tareknaous/bert2bert-empathetic-response-msa
chat_tokenizer = AutoTokenizer.from_pretrained("tareknaous/bert2bert-empathetic-response-msa")
chat_model = EncoderDecoderModel.from_pretrained("tareknaous/bert2bert-empathetic-response-msa")
chat_model.eval()
```

```python
from arabert.preprocess import ArabertPreprocessor
arabert_prep = ArabertPreprocessor(model_name="bert-base-arabert", keep_emojis=False)
```

```python
from arabert.preprocess import ArabertPreprocessor
model_name = "bert-base-arabert"
arabert_prep = ArabertPreprocessor(model_name=model_name)
```

```python
def ask(text):
    # Check if the user input matches any pattern in NLTK chatbot pairs
    nltk_response = nltk_chatbot.respond(text)
    if nltk_response:
        return nltk_response

    response, score = findSimilarity(text)
    if(score > 0.5):
      return response 

    text_clean = arabert_prep.preprocess(text)
    inputs = chat_tokenizer.encode_plus(text_clean,return_tensors='pt')
    outputs = chat_model.generate(input_ids = inputs.input_ids,
                    attention_mask = inputs.attention_mask,
                    do_sample = True,
                    top_k = 3,
                    top_p = 0.9,
                    temperature = 0.3)
    preds = chat_tokenizer.batch_decode(outputs) 
    response = str(preds)
    response = response.replace("\'", '')
    response = response.replace("[[CLS]", '')
    response = response.replace("[SEP]]", '')
    response = str(arabert_prep.desegment(response))
    return response
```
