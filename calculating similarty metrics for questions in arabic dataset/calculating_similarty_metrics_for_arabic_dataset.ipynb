```python
from sentence_transformers import SentenceTransformer, util

# Load the pre-trained model
sentence_similarity = SentenceTransformer('symanto/sn-xlm-roberta-base-snli-mnli-anli-xnli')
```

```python
#@title loading df variable
import pandas as pd
from google.colab import drive
import os, glob

drive.mount('/content/drive')
file_name="arabic-dataframe with encoding.csv"#@param {type:"string"}
current_colab_name="calculating similarty metrics for arabic dataset.ipynb"#@param {type:"string"}

if 'my_path' not in locals():
    my_path = glob.glob(os.getcwd() + '/**/' + current_colab_name, recursive=True)
    my_path = [my_i for my_i in my_path if len(my_i) in [max([len(my_i) for my_i in my_path])]][0]
df = pd.read_csv(os.path.dirname(my_path)+"/"+file_name)
```

```python
df
```

```python
subset_df = df[['question', 'tags']].copy()
```

```python
subset_df
```

```python
def select_top_3_entries(group):
    return group.head(3)
```

```python
subset_df = subset_df.groupby('tags', group_keys=False).apply(select_top_3_entries)
```

```python
import json
subset_json = subset_df.to_json(orient='records')
with open('subset_data.json', 'w') as json_file:
    json.dump(json.loads(subset_json), json_file, indent=4)
```

```python
# Extract the "question" column from the DataFrame
questions = subset_df['question']

# Define the file path for the text file
file_path = 'questions.txt'

# Write the questions to the text file
with open(file_path, 'w') as file:
    for question in questions:
        file.write(question + '\n')
```

```python
paraphrased_questions = []
file_path = 'paraphrased questions.txt'
with open(file_path, 'r') as file:
    for line in file:
        paraphrased_questions.append(line.strip())

# Add the paraphrased questions as a new column in the DataFrame
subset_df['paraphrased_question'] = paraphrased_questions
```

```python
# Calculate sentence embeddings for the original questions
subset_df['question_embedding'] = subset_df['question'].apply(lambda x: sentence_similarity.encode(x, convert_to_tensor=True))

# Calculate sentence embeddings for the paraphrased questions
subset_df['paraphrased_embedding'] = subset_df['paraphrased_question'].apply(lambda x: sentence_similarity.encode(x, convert_to_tensor=True))
```

```python
import torch

# Calculate cosine similarity scores
subset_df['cosine_similarity'] = subset_df.apply(lambda row: util.pytorch_cos_sim(row['question_embedding'], row['paraphrased_embedding']).item(), axis=1)
```

```python
average_similarity = subset_df['cosine_similarity'].mean()
max_similarity = subset_df['cosine_similarity'].max()
min_similarity = subset_df['cosine_similarity'].min()
std_similarity = subset_df['cosine_similarity'].std()
high_similarity_count = len(subset_df[subset_df['cosine_similarity'] > 0.8])
low_similarity_count = len(subset_df[subset_df['cosine_similarity'] < 0.2])

# Print metrics
print(f"Average Cosine Similarity: {average_similarity:.4f}")
print(f"Maximum Cosine Similarity: {max_similarity:.4f}")
print(f"Minimum Cosine Similarity: {min_similarity:.4f}")
print(f"Standard Deviation of Cosine Similarity: {std_similarity:.4f}")
print(f"Count of Entries with High Similarity (> 0.8): {high_similarity_count}")
print(f"Count of Entries with Low Similarity (< 0.2): {low_similarity_count}")
```

```python
import matplotlib.pyplot as plt

plt.hist(subset_df['cosine_similarity'], bins=20, range=(0, 1))
plt.xlabel('Cosine Similarity')
plt.ylabel('Frequency')
plt.title('Histogram of Cosine Similarity Scores')
plt.show()
```
