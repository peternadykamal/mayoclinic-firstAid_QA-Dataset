```python
import os
import xml.etree.ElementTree as ET
import pandas as pd

# Create an empty list to hold the data
data = []

# Loop through all XML files in the current directory
for filename in os.listdir("."):
    if filename.endswith(".xml"):
        # Parse the XML file
        tree = ET.parse(filename)
        root = tree.getroot()

        # Loop through all Question elements in the file
        for question in root.findall("Question"):
            # Get the QAquestion and QAanswer elements and extract their text
            question_text = question.find("QApair/QAquestion").text
            answer_text = question.find("QApair/QAanswer").text

            # Add the question and answer to the data list
            data.append({"question": question_text, "answer": answer_text})

# Convert the data list to a DataFrame
df = pd.DataFrame(data)
```

```python
import re
import string
import nltk
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('arabic'))

# remove Arabic question words from the stopwords
question_words = ['من', 'ماذا', 'كيف', 'أين', 'متى', 'لماذا', 'ممن', 'كم', 'أي', 'هل', 'أيّان']
negating_words = ['لا', 'لاسيما', 'ليس', 'ليست', 'لست', 'لستُ', 'لستِ', 'لستم', 'لستنّ', 'لستما', 'ليسوا', 'لستُم', 'لستُن', 'ليسون', 'إنما', 'بل', 'ولكن', 'إذن', 'إن', 'لكن', 'عكس', 'مع', 'ما عدا', 'إلا']
for word in question_words:
    if word in stop_words:
      stop_words.remove(word)

for word in negating_words:
    if word in stop_words:
        stop_words.remove(word)
```

```python
#@title used this cell to calcalute the most common words in the dataset

# concatenate all answers into a single string
text = ' '.join(df['answer'].tolist())
text = ' '.join(df['question'].tolist())

# tokenize the answers string
tokens = nltk.word_tokenize(text)

# remove stop words and non-alphabetic tokens
tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]

# calculate frequency distribution of tokens
fdist = FreqDist(tokens)

# get 10 most common words
most_common_words = fdist.most_common(16)

# print the most common words
print(most_common_words)

most_common_words = [w[0] for w in most_common_words]
```

```python
most_common_words = ['الله', 'يكون', 'وبعد', 'عليكم', 'يجب', 'السلام', 'الدم', 'بسم', 'الرحمن', 'الإجابــة', 'وبركاته', 'الرحيم', 'يمكن', 'السؤال', 'ورحمة', 'وبعد']
```

```python
#@title preprocess_arabic_text function
def preprocess_arabic_text(text):
    # convert text to lowercase
    text = text.lower()

    # remove punctuation and non-Arabic characters
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text)

    # tokenize the text
    tokens = word_tokenize(text)

    # remove stop words
    filtered_words = []
    for word in tokens:
      if word not in stop_words and word not in most_common_words:
          filtered_words.append(word)

    # join the tokens back into a string
    text = ' '.join(filtered_words)

    return text
```

```python
#@title remove_stopwords function
def remove_stopwords(text):
    text = text.translate(str.maketrans("", "", string.punctuation))
    words = text.split()
    filtered_words = []
    for word in words:
        if word not in stop_words and word not in most_common_words:
            filtered_words.append(word)
    return ' '.join(filtered_words)
```

```python
#@title remove_most_common_words function
def remove_most_common_words(text):
    words = text.split()
    filtered_words = []
    for word in words:
        if word not in most_common_words:
            filtered_words.append(word)
    return ' '.join(filtered_words)
```

```python
import re
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import string
import torch

device_name = "cuda" if torch.cuda.is_available() else "cpu"
model_name = "csebuetnlp/mT5_multilingual_XLSum"
tokenizer = AutoTokenizer.from_pretrained(model_name,device=device_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
device = torch.device(device_name)
model.to(device)
```

```python
def summarize(text):
    WHITESPACE_HANDLER = lambda k: re.sub('\s+', ' ', re.sub('\n+', ' ', k.strip()))

    #text preprocessing
    #preprocessed_text = remove_stopwords(text) #used for answer column
    preprocessed_text = preprocess_arabic_text(text) #used for question column
    txtLength = len(preprocessed_text.split())

    #if txtLength < 20: return text #used for answer column
    if txtLength < 35: return remove_stopwords(text) #used for question column

    input_ids = tokenizer(
        [WHITESPACE_HANDLER(preprocessed_text)],
        return_tensors="pt",
        padding="max_length",
        truncation=True,
        max_length=512
    )["input_ids"]
    input_ids = input_ids.to(device)
    
    output_ids = model.generate(
        input_ids=input_ids,
        max_length=txtLength,
        min_length=int(txtLength * 0.7),
        no_repeat_ngram_size=2,
        num_beams=4
    )[0]

    summary = tokenizer.decode(
        output_ids,
        skip_special_tokens=True,
        clean_up_tokenization_spaces=False
    )

    return summary
```
